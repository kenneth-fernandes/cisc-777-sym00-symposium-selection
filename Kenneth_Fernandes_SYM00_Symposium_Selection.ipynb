# SYM00 — Symposium / Venue Selection

## Adaptive KV-Cache Precision and Layout Selection for Long-Context LLM Inference

**Kenneth Peter Fernandes** · CISC 777 · June 29, 2026

---

My paper is a **machine-learning *systems*** contribution — a characterization study, predictive model, and adaptive Triton-kernel runtime for choosing KV-cache precision (INT4/INT8/FP16) and block layout at decode time. It belongs in the **efficient-ML-systems / GPU-inference** community, so the search below is *systems-leaning*.

**Ranking basis:** conferences by [CORE rank](https://portal.core.edu.au/conf-ranks/) + reputation (A\*/top = high, B = medium, workshop = low); journals by Clarivate impact factor + JCR quartile.

## Summary & strategy

| Tier | Conference | Journal |
|------|-----------|---------|
| High | MLSys | IEEE TPDS (Q1, A\*) |
| Medium | PACT (CORE B) | ACM TACO (Q2, A) |
| Low | EuroMLSys workshop | TMLR (free) |

Conference totals cluster near \$1,600–1,750 because **location, not prestige, drives cost** — the low-rank EuroMLSys is expensive only because it is overseas. Journals invert the usual intuition: the **highest-ranked** TPDS can be **\$0** via subscription, while mid-tier TACO carries the largest mandatory APC.

**Plan:** present early results at **EuroMLSys/ENLSP**, target **MLSys** for the full paper, and use **TPDS** (free route) or **TACO** for a journal version, with **TMLR** as a zero-cost fallback.
